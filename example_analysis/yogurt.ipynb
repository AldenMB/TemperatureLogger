```python
with open('2021_1_1_0_0.txt', 'r') as f:
    next(f)
    logs = list(f.readlines())
```

```python
timestamps, yogurt, ambient = zip(*(line.split() for line in logs))
yogurt = [float(y) for y in yogurt]
ambient = [float(a) for a in ambient]
def stamp2secs(stamp):
    hours, minutes, seconds = (int(t) for t in stamp.split(':'))
    return seconds + 60*(minutes+60*hours)
seconds = [stamp2secs(s) for s in timestamps]

yogurt = yogurt[5:]
seconds = seconds[5:]
ambient = ambient[5:]
```

```python
import matplotlib.pyplot as plt
plt.plot(seconds, yogurt)
plt.plot(seconds, ambient)
plt.show()
```

```python
import numpy as np
```

```python
t = np.array(seconds)
a = np.array(ambient)
y = np.array(yogurt)
```

```python
a.max(), a.min()
```

```python
z = np.log(y-a)
plt.plot(t, z)
plt.show()
```

```python
m, b = np.polyfit(t, z, 1)
fit = b+m*t
plt.plot(t, z)
plt.plot(t, fit)
plt.show()
```

```python
plt.plot(t, z-fit)
plt.show()
```

```python
plt.plot(t, y)
plt.plot(t, np.exp(fit)+a)
plt.show()
```
